```python
import pandas as pd
```

```python
location = "airport-codes.csv"
airport_data = pd.read_csv(location)
print(airport_data.shape)
airport_data.head()
```

```python
us_airport_data = airport_data.loc[airport_data["iso_country"] == "US"]
print(us_airport_data.shape)
us_airport_data.head()
```

```python
airport_trim = us_airport_data.loc[us_airport_data["iata_code"].notnull()]
airport_trim.head()
```

```python
airport_trim = airport_trim[["iata_code", "name", "coordinates", "iso_region"]]
airport_trim.head()
```

```python
airport_trim.to_csv("cleaned_airport_list.csv")
```
